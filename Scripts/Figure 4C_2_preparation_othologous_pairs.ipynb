```python
import pandas as pd
import os
```

```python
# path
path_in_cf = '01_merge_bed/cf3'
path_in_hg = '01_merge_bed/hg38'
path_in_mm = '01_merge_bed/mm10'
path_out_hg = '02_ortho_bed_w2species/cf3_hg38'
path_out_mm = '02_ortho_bed_w2species/cf3_mm10'
```

```python
ls_in_cf = sorted(os.listdir(path_in_cf))
ls_in_hg = sorted(os.listdir(path_in_hg))
ls_in_mm = sorted(os.listdir(path_in_mm))
```

```python
ls_in_cf
```

```python
## Human
for file in ls_in_hg:
	tissue = file.split('_')[0]
	mark = file.split('_')[2].split('.')[0]
	print(f'# {tissue}, {mark}')
	
	# Other species
	pd_in = pd.read_csv(f'{path_in_hg}/{file}', sep='\t', header=None, index_col=None)
	pd_in_label = pd_in[3]
	# Canine
	pd_in_cf = pd.read_csv(f'{path_in_cf}/{tissue}_cf3_{mark}.bed', sep='\t', header=None)
	pd_in_cf.index = pd_in_cf[3]
	
	# Extract row
	pd_in_cf_orth = pd_in_cf.loc[pd_in_label]
	print(f'hg, cf, ortho: {pd_in_label.shape[0]} / {pd_in_cf.shape[0]}/ {pd_in_cf_orth.shape[0]}')
	
	# Write ortho bed
	pd_in_cf_orth.to_csv(f'{path_out_hg}/{tissue}_cf3_{mark}_ortho_hg38.bed', sep='\t', header=None, index=None)
```

```python
## Mouse
for file in ls_in_mm:
	tissue = file.split('_')[0]
	mark = file.split('_')[2].split('.')[0]
	print(f'# {tissue}, {mark}')
	
	# Other species
	pd_in = pd.read_csv(f'{path_in_mm}/{file}', sep='\t', header=None, index_col=None)
	pd_in_label = pd_in[3]
	# Canine
	pd_in_cf = pd.read_csv(f'{path_in_cf}/{tissue}_cf3_{mark}.bed', sep='\t', header=None)
	pd_in_cf.index = pd_in_cf[3]
	
	# Extract row
	pd_in_cf_orth = pd_in_cf.loc[pd_in_label]
	print(f'hg, cf, ortho: {pd_in_label.shape[0]} / {pd_in_cf.shape[0]}/ {pd_in_cf_orth.shape[0]}')
	
	# Write ortho bed
	pd_in_cf_orth.to_csv(f'{path_out_mm}/{tissue}_cf3_{mark}_ortho_mm10.bed', sep='\t', header=None, index=None)
```
